```python
import tkinter as tt
from tkinter import ttk

window=tt.Tk()
window.title("Automated Data Backup & Recovery System")
window.geometry("1000x650")
window.configure(bg="#146ca4")

#header code
head=tt.Label(
    window,
    text="AUTOOMATED DATA BCAKUP & RECOVERY MANAGMENT SYSTEM",
    font=("Times New Roman",18,"bold"),
    bg="#146ca4",
    fg="white",
    pady=15
)
head.pack(fill="x")


#left side

menu_bar=tt.Frame(window,bg="#52b7e9",width=200)
menu_bar.pack(side="left", fill="y")

buttons=[
    "Dashboard",
    "Create Backup",
    "backup History",
    "Restore Files",
    "Backup System",
    "Logs",
    "Settings",
    "exit"
] 

for items in buttons:
    bt=tt.Button(
        menu_bar,text=items,font=("Arial",12),
        bg="#34495e",
        fg="white",
        padx=10,
        pady=10
    )
    bt.pack(fill="x",padx=5,pady=2)


#Main area
main=tt.Frame(window,bg="white")
main.pack(fill="both",expand=True)

title=tt.Label(
    main,
    text="Dashboard",
    font=("Arial",16,"bold"),
    bg="white"
)
title.pack(pady=15)

#information detail
card=tt.Frame(main, bg="white")
card.pack(pady=10)

def creates_card(parent,title,value):
    frame=tt.Frame(parent,bg="#024950",width=180,height=100)
    frame.pack(side="left", padx=15)


    tt.Label(
        frame,
        text=title,
        font=("Arial",18,"bold"),
        bg="#3498db",
        fg="white",
    ).pack()

    tt.Label(
        frame,
        text=value,
        font=("Arial",18),
        bg="#3498db",
        fg="white"
    ).pack()

creates_card(card,"TOTAL Backups","10")
creates_card(card,"Sucessful","14")
creates_card(card,"Failed","1")
creates_card(card,"Storage USED","4.0 Gb")



#bottoms button
bottom=tt.Frame(main, bg="white")
bottom.pack(pady=15)

tt.Button(bottom, text="Create Backup", width=18, bg="orange", fg="white").pack(side="left", padx=10)

tt.Button(bottom, text="Refresh", width=18,bg="pink",fg="white").pack(side="left", padx=10)

tt.Button(bottom, text="Restore file",width=18 ,bg="red",fg="white").pack(side="left", padx=10)

window.mainloop()

```
